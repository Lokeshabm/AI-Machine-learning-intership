```python
import matplotlib.pyplot as plt
moths=["jan","fub","mar","apr","may","june","july","agt","spt","oct","nou","dec"] 
energy=[1200,2300,2000,3100,1250,2302,2000,4000,5000,8000,7000,9000]
plt.plot(moths,energy,marker='+',color='b',linestyle=':')
plt.xlabel("months")
plt.ylabel('energy(kwh)')
plt.title("monthly energy comsumption")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
months=["jan","fub","mar","apr","may","june","july","agt","spt","oct","nou","dec"] 
energy=[1200,2300,2000,3100,1250,2302,2000,4000,5000,8000,7000,9000]
plt.bar(moths,energy,color='b',linestyle=':')
plt.xlabel("months")
plt.ylabel('energy(kwh)')
plt.title("monthly energy comsumption")
plt.show()
```

```python
plt.pie(energy,labels=months,autopct='%1.1f%%',colors=["gold","lightblue","lightgreen","coral"])
plt.title("energy consumption share by source")
plt.show
```
